# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func, desc
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
# engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Inspect the measurement data file
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
```

```python
# Inspect the station data file
inspector = inspect(engine)
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c["type"])
```

# Precipitation Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Find the most recent date
most_recent_date = session.query(Measurement.date).order_by(desc(Measurement.date)).first() 
print(most_recent_date)
```

```python
# Establish the date range to be used
max_date = dt.date(2017, 8, 23)
min_date = max_date - dt.timedelta(days=365) 
print(f"The date range is {min_date} to {max_date}")
```

```python
# Perform a query to retrieve the data and precipitation scores
ly_results = session.query(Measurement.id, Measurement.station, Measurement.date, Measurement.prcp, Measurement.tobs).\
    filter(func.strftime('%Y-%m-%d', Measurement.date) >= min_date)
print(ly_results)
ly_results.all()
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
year_df = pd.DataFrame(ly_results, columns = ['pk_id', 'station', 'date', 'prcp', 'tobs'])
year_df.head()
```

```python
# Replace precipitation NaNs with 0s
year_df = year_df.fillna(0)
year_df.head()
```

```python
# Replace precipitation NaNs with 0s
precipitation_df = year_df[["date", "prcp"]]
precipitation_df.head()
```

```python
# Group by date and sort by date
precipitation_df = precipitation_df.groupby(["date"]).sum().sort_values(by='date').reset_index()
precipitation_df.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
plt.plot(precipitation_df["date"], precipitation_df["prcp"])
plt.title("Precipitation by Day")
plt.ylabel("Precipitation")
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```

```python
# Calcualte the summary statistics for the precipitation data
mean_prcp = round(np.mean(precipitation_df["prcp"]),2)
median_prcp = round(np.median(precipitation_df["prcp"]), 2)
var_prcp = round(np.var(precipitation_df["prcp"], ddof=0),2)
sd_prcp = round(np.std(precipitation_df["prcp"], ddof=0),2)

print(f"The average precipitation is {mean_prcp}")
print(f"The median precipitation is {median_prcp}")
print(f"The precipitation variance is {var_prcp}")
print(f"The precipitation standard deviation is {sd_prcp}")
```

# Station Analysis

```python
# Design a query to show how many stations are available in this dataset?
station_count_query = session.query(Station.station).distinct().all()
print(f"There are {len(station_count_query)} total stations")
```

```python
active_stations_query = session.query(Measurement.station).distinct().all()
print(f"There are {len(active_stations_query)} active stations")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
most_active_stat_query = session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.station).desc())
print(most_active_stat_query)
most_active_stat_query.all()
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
station_summary = session.query(Measurement.station, func.min(Measurement.tobs), func.avg(Measurement.tobs),\
    func.max(Measurement.tobs)).\
    filter(Measurement.station == 'USC00519281').all()
print("Min, Average, and Max temp for the most active station USC00519281 is:")
station_summary
# (54.0, 85.0, 71.66378066378067)]
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
ly_station_temp = session.query(Measurement.date, func.avg(Measurement.tobs)).\
    filter(func.strftime('%Y-%m-%d', Measurement.date) >= min_date).\
    filter(Measurement.station == 'USC00519281').\
    group_by(Measurement.date).\
    order_by((Measurement.date).desc())
print(ly_station_temp)
ly_station_temp.all()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
